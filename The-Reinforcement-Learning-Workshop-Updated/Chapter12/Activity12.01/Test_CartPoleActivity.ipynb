```python
import unittest
import import_ipynb
import numpy as np
import gym

class TestCartPoleActivity(unittest.TestCase):
    def setUp(self):
        import CartPoleActivity
        self.exercises = CartPoleActivity
    
    def test_ga_algo(self):
        env = gym.make('CartPole-v0') 
        self.assertEqual(len(self.exercises.ga_algo(env, 5, 5)),4)
    
    def test_params(self):
        env = gym.make('CartPole-v0') 
        p= self.exercises.ga_algo(env, 5, 5)
        self.assertEqual(len(self.exercises.params(p)),len(self.exercises.ga_algo(env, 5, 5)))
        
    def test_nnmodel(self):
        env = gym.make('CartPole-v0') 
        p = self.exercises.ga_algo(env, 5, 5)
        x,y,z,q = self.exercises.params(p)
        s = env.reset()
        a = self.exercises.nnmodel(s,x,y,z,q)
        self.assertEqual(a>=0, True)
                
```

```python
suite = unittest.TestLoader().loadTestsFromTestCase(TestCartPoleActivity)
unittest.TextTestRunner(verbosity=2).run(suite)
```
