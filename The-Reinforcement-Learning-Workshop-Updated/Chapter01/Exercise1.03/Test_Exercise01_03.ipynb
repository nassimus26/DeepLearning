```python
import import_ipynb
import numpy as np
import unittest


class TestExercise01_03(unittest.TestCase):
    def setUp(self):
        import Exercise01_03

        self.exercise = Exercise01_03
    
    def test_space(self):
        """
        Assert that the shape of space is correct (256, 256, 3)
        Assert that the dtype of space is correct: np.uint8
        Assert that the low value of space is correct: zeros of the correct shape
        Assert that the high value of space is correct: 255 of the correct shape
        """
        self.assertTrue(self.exercise.space.shape == (256, 256, 3)) 
        self.assertTrue(self.exercise.space.dtype == np.uint8) 
        np.testing.assert_array_equal(self.exercise.space.low, np.zeros((256, 256, 3)))
        np.testing.assert_array_equal(self.exercise.space.high, np.ones((256, 256, 3))*255)
    
    def test_space_temporal(self):
        """
        Assert that the shape of space_temporal is correct (4, 256, 256, 3)
        Assert that the dtype of space_temporal is correct: np.uint8
        Assert that the low value of space_temporal is correct: zeros of the correct shape
        Assert that the high value of space_temporal is correct: 255 of the correct shape
        """
        self.assertTrue(self.exercise.space_temporal.shape == (4, 256, 256, 3)) 
        self.assertTrue(self.exercise.space_temporal.dtype == np.uint8) 
        np.testing.assert_array_equal(self.exercise.space_temporal.low, np.zeros((4, 256, 256, 3)))
        np.testing.assert_array_equal(self.exercise.space_temporal.high, np.ones((4, 256, 256, 3))*255)
```

```python
suite = unittest.TestLoader().loadTestsFromTestCase(TestExercise01_03)
unittest.TextTestRunner(verbosity=2).run(suite)
```
